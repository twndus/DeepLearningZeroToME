```python
import numpy as np
import pandas as pd

import tensorflow as tf
```

입력 데이터 정의

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.]
```

1. implement matrix-based regression graph

```python
learning_rate = 0.00001

w1 = tf.Variable(initial_value=(-1,), shape=(1,), dtype='float32')
w2 = tf.Variable(initial_value=(-1,), shape=(1,), dtype='float32')
w3 = tf.Variable(initial_value=(-1,), shape=(1,), dtype='float32')
b = tf.Variable(initial_value=(-1,), shape=(1,), dtype='float32')

@tf.function
def cost():
    hypothesis = x1_data*w1 + x2_data*w2 + x3_data*w3 + b
    loss = tf.reduce_mean(tf.square(hypothesis-y_data))
    return loss

sgd = tf.optimizers.legacy.SGD(learning_rate=learning_rate)
```

```python
for i in range(2001):
    sgd.minimize(cost, var_list=[w1,w2,w3,b], tape=tf.GradientTape())
    hypothesis = x1_data*w1 + x2_data*w2 + x3_data*w3 + b
    print(f"epoch: {i}, cost: {cost().numpy()}, prediction: {hypothesis}")
```

1-2.

```python
x_data = np.array([x1_data, x2_data, x3_data], dtype='float32').T
x_data.shape
```

```python
learning_rate = 0.00001

W = tf.Variable(initial_value=(-1,-1,-1), shape=((3,)), dtype='float32')
b = tf.Variable(initial_value=(-1,), shape=(1,), dtype='float32')

@tf.function
def cost():
    hypothesis = tf.reduce_sum(tf.multiply(x_data, W) + b, axis=1)
    loss = tf.reduce_mean(tf.square(hypothesis-np.expand_dims(y_data, -1)))
    return loss

sgd = tf.optimizers.legacy.SGD(learning_rate=learning_rate)
```

```python
for i in range(2001):
    sgd.minimize(cost, var_list=[W,b], tape=tf.GradientTape())
    hypothesis = tf.reduce_sum(tf.multiply(x_data, W) + b, axis=1)
    print(f"epoch: {i}, cost: {cost().numpy()}, prediction: {hypothesis}")
```
